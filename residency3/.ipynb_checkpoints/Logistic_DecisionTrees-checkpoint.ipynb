# Classification vs. Regression
The aim of this project is to predict how likely a student is to pass. Which type of supervised learning problem is this, classification or regression? Why?
Answer:
This project is a classification supervised learning problem because the variable to predict, i.e. if a student graduates or fails to graduate, is categorical. On this case this a dichotomous categorical variable where the only two possible values are "pass" or "fail".

# Question 6 - Encode all categorical variables to numerical (0.5 points)

Take the list of categorical attributes(from the above result) and convert them into neumerical variables. After that, print the head of dataframe and check the values.

**Hint:** check **sklearn LabelEncoder()**

#### Answer:

```python
from sklearn import preprocessing
#le = preprocessing.LabelEncoder()

for c in stud.columns:
        if stud[c].dtype == 'object':
            lbl = preprocessing.LabelEncoder()
            lbl.fit(list(stud[c].values))
            stud[c] = lbl.transform(list(stud[c].values))
```

# Question 7 - Convert the continuous values of grades into classes (1 point)

*Consider the values in G1, G2 and G3 with >= 10 as pass(1) and < 10 as fail(0) and encode them into binary values. Print head of dataframe to check the values.*

#### Answer:

```python
stud.shape
```

```python
stud.head()
```

```python
stud["G1"] = stud["G1"].map(lambda
         x: 0 if (x<10)
         else 1)
stud["G2"] = stud["G2"].map(lambda
         x: 0 if (x<10)
         else 1)
stud["G3"] = stud["G3"].map(lambda
         x: 0 if (x<10)
         else 1)
```

```python
stud.head()
```

# Information about the dataset

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. 

Attribute Information
1. ID - ID number of the image
2. Clump Thickness - 1 to 10
3. Cell Size - 1 to 10
4. Cell Shape - 1 to 10
5. Marginal Adhesion - 1 to 10
6. Single Epethelial cell size - 1 to 10
7. Bare nuclei - 1 to 10 
8. Normal Nucleoli - 1 to 10
9. Bland Chromatin - 1 to 10
10. Mitosis - 1 to 10
 
Class - Dependent Variable - 2 for benign 4 for malignant

### Divide the dataset into feature set and target set

```python
X=breast.drop(['Class'],axis=1)
y=breast['Class']
```

### Divide the Training and Test sets in 70:30 

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape
```

### Model 1 : Implement a default Decision Tree model with no attributes

```python
from sklearn.tree import DecisionTreeClassifier

reg_dt_model = DecisionTreeClassifier()
reg_dt_model.fit(X_train, y_train)
```

```python
y_predict = reg_dt_model.predict(X_test)
```

```python
reg_dt_model.score(X_test , y_test)
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_test, y_predict)
```

### Check the feature importances of the Decision Tree model

```python
pd.DataFrame(reg_dt_model.feature_importances_, columns = ["Imp"], index = X_test.columns)
```

### Check the training & test accuracy of model 1 

```python
reg_dt_model.score(X_test , y_test)
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_test, y_predict)
```

### Model 2: Implement a decision tree with depth of the tree = 10 and leaf nodes = 5

```python
reg_dt_model = DecisionTreeClassifier(criterion = 'entropy', max_leaf_nodes=5,max_depth = 10)
reg_dt_model.fit(X_train, y_train)
```

```python
y_predict = reg_dt_model.predict(X_test)
```

### Check the training & test accuracy of model 2

```python
y_predict = reg_dt_model.predict(X_test)
reg_dt_model.score(X_test , y_test)
```

### Print the confusion matrix of model 1 and model 2

```python
from sklearn import metrics
metrics.confusion_matrix(y_test, y_predict)
```
